# PandasGUI

## About

PandasGUI is a GUI for viewing, plotting and analyzing Pandas DataFrames.

## Features

- View DataFrames and Series (with MultiIndex support)
- Interactive plotting
- Filtering
- Statistics summary
- Data editing and copy/paste
- Import CSV files with drag & drop
- Search toolbar

## Usage

```python
# Create and view a simple DataFrame

import pandas as pd
from pandasgui import show
```

```python
df = pd.DataFrame({'a':[1, 2, 3], 'b':[4, 5, 6], 'c':[7, 8, 9]})
show(df)
```

```python
# PandasGUI comes with sample datasets that will download on first use.
# You can also import all_datasets which is a dictionary of all sample 
# datasets like Dict[str, DataFrame]
from pandasgui import show
from pandasgui.datasets import pokemon, titanic, all_datasets
show(pokemon, titanic)
```

```python
show(**all_datasets)
```
